```python
import ezfit
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def line(x, m, b):
    return m * x + b


test_data = pd.DataFrame(
    {
        "x": np.linspace(0, 10, 100),
        "y": np.linspace(0, 10, 100) + np.random.normal(0, 1, 100),
        "y_err": np.random.poisson(1, 100) + .001
    }
)
model, ax = test_data.fit(line, "x", "y")
print(model)
plt.show()
```
