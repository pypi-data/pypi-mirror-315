```python
import semchunk as old_semchunk
import sys
sys.path.append('D:/workspace/semchunk/')
import src.semchunk.semchunk as semchunk
from importlib import reload
import nltk
import tiktoken
import time
import semantic_text_splitter

nltk.download('gutenberg')
gutenberg = nltk.corpus.gutenberg
encoding = tiktoken.encoding_for_model('gpt-4')

def _token_counter(text: str) -> int:
    return len(encoding.encode(text))

samples: list[str] = [gutenberg.raw(fileid) for fileid in gutenberg.fileids()]

chunk_size = 512
reload(semchunk)
```

```python
import array
import numpy as np
from itertools import accumulate
splits = samples[0].split('\n')

def acum(texts: list[str]) -> list[int]:
    cum = list(accumulate([len(split) for split in texts], initial=0))
    return cum + [cum[-1]]

def a1(texts: list[str]) -> list[int]:
    cum = array.array('I', [0])
    last = 0
    
    for text in texts:
        last += len(text)
        cum.append(last)
    
    cum.append(cum[-1])
    
    return cum

def a2(texts: list[str]) -> list[int]:
    cum = array.array('I', [0])
    last = 0
    cum.extend([(last := last + len(text)) for text in texts])
    cum.append(cum[-1])
    
    return cum

def a3(texts: list[str]) -> list[int]:
    cum = [0] + list(accumulate([len(split) for split in texts]))
    return cum + [cum[-1]]

%timeit acum(splits)
%timeit a3(splits)
# %timeit a2(splits)
```
